```python
import cmath
a= float(input("a:"))
b= float(input("b:"))
c= float(input("c:"))
delta = b**2-4*a*c
if delta >0:
  x1= (-b+cmath.sqrt(delta))/(2*a)
  x2= (-b-cmath.sqrt(delta))/(2*a)
  print("raices reales:",x1,x2)
elif delta ==0:
    x= -b/(2*a)
    print("raices reales:",x)
```
